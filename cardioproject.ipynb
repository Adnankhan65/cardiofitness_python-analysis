**BUSINESS QUESTIONS**

```python
##How many models does store have?
df['Product'].nunique()
```

```python
## Which is most sold Model? 
df['Product'].value_counts()
```

```python
# Plot
plt.figure()
df['Product'].value_counts().plot(kind='bar')
plt.title('Product Distribution')
plt.show()
```

```python
##Are Male customers buying treadmill more than female customers? 
print(df['Gender'].value_counts())
```

```python
plt.figure()
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.show()
```

```python
##What is the Income, Age, Education of people buying treadmill.? 
print(df[['Age','Income','Education']].describe())
```

```python
plt.figure()
df['Age'].hist()
plt.title('Age Distribution')
plt.show()

plt.figure()
df['Income'].hist()
plt.title('Income Distribution')
plt.show()
```

```python
df.columns
```

```python
##How many days and miles customer expect to run on treadmill? 
df[['Usage','Miles']].describe()
```

```python
plt.figure()
df['Usage'].hist()
plt.title('Usage per Week')
plt.show()

plt.figure()
df['Miles'].hist()
plt.title('Miles Distribution')
plt.show()
```

```python
 #What is the self-rated fitness of customers buying treadmill.? 
df['Fitness'].value_counts()
```

```python
plt.figure()
df['Fitness'].value_counts().plot(kind='bar')
plt.title('Fitness Levels')
plt.show()
```

```python
#Is married customer buying Treadmill more than Single customers? 
print(df['MaritalStatus'].value_counts())
```

```python
plt.figure()
df['MaritalStatus'].value_counts().plot(kind='bar')
plt.title('MaritalStatus')
plt.show()
```

```python
 #Is there any relation between Income and model.? 
df.groupby('Product')['Income'].mean()
```

```python
#Age vs Model
df.groupby('Product')['Age'].mean()
```

```python
#fitness vs model
df.groupby('Product')['Fitness'].mean()
```

```python
#Education vs Model
df.groupby('Product')['Education'].mean()
```

```python
#Gender vs Model
pd.crosstab(df['Product'], df['Gender'])
```

```python
# Marital Status vs Product
print(pd.crosstab(df['Product'], df['MaritalStatus']))
```

```python
df['Age'].value_counts()
df['Age'].min()
df['Age'].max()
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[18, 25, 35, 45, 50],
    labels=['18-25', '26-35', '36-45', '46-50']
)
```

```python
pd.crosstab(df['Age_Group'], df['Product'])
```

```python
pd.crosstab(df['Age_Group'], df['Product']).plot(kind='bar')
plt.title('Gender Distribution Across products')
plt.show()
```
